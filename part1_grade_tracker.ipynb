```python
# ==========================================
# TASK 1: Data Parsing & Cleaning
# ==========================================

raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma", "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair ", "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA", "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ", "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

cleaned_students = []

for student in raw_students:
    name = student["name"].strip().title()
    roll = int(student["roll"])
    marks = list(map(int, student["marks_str"].split(",")))

    if all(word.isalpha() for word in name.split()):
        print("✓ Valid name")
    else:
        print("✗ Invalid name")

    cleaned_students.append({
        "name": name,
        "roll": roll,
        "marks": marks
    })

# Profile cards
for s in cleaned_students:
    print("\n========================")
    print(f"Student : {s['name']}")
    print(f"Roll No : {s['roll']}")
    print(f"Marks   : {s['marks']}")
    print("========================")

# Special output for roll 103
for s in cleaned_students:
    if s["roll"] == 103:
        print("\nSpecial Output:")
        print(s["name"].upper())
        print(s["name"].lower())
```

```python
# ==========================================
# TASK 2: Marks Analysis (Modified for Colab)
# ==========================================

student_name = "Ayesha Sharma"
subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]

def get_grade(m):
    if m >= 90:
        return "A+"
    elif m >= 80:
        return "A"
    elif m >= 70:
        return "B"
    elif m >= 60:
        return "C"
    else:
        return "F"

# Print grades
print("\nSubject-wise Grades:")
for i in range(len(subjects)):
    print(subjects[i], "-", marks[i], "-", get_grade(marks[i]))

# Calculations
total = sum(marks)
avg = round(total / len(marks), 2)

print("\nTotal:", total)
print("Average:", avg)

# Highest & Lowest
max_marks = max(marks)
min_marks = min(marks)

print("Highest:", subjects[marks.index(max_marks)], max_marks)
print("Lowest:", subjects[marks.index(min_marks)], min_marks)

# 🔥 Simulated input (instead of input())
print("\nSimulating additional subject input...")

extra_subjects = ["Biology", "done"]
extra_marks = [85]

i = 0
while True:
    subject = extra_subjects[i]
    print("Enter subject:", subject)

    if subject.lower() == "done":
        break

    mark = extra_marks[i]
    print("Enter marks:", mark)

    if not (0 <= int(mark) <= 100):
        print("Invalid marks")
        continue

    subjects.append(subject)
    marks.append(int(mark))

    i += 1

print("\nUpdated Average:", round(sum(marks)/len(marks), 2))

```

```python
# ==========================================
# TASK 3: Class Performance Summary
# ==========================================

class_data = [
    ("Ayesha Sharma", [88, 72, 95, 60, 78]),
    ("Rohit Verma", [55, 68, 49, 72, 61]),
    ("Priya Nair", [91, 85, 88, 94, 79]),
    ("Karan Mehta", [40, 55, 38, 62, 50]),
    ("Sneha Pillai", [75, 80, 70, 68, 85]),
]

results = []

for name, marks in class_data:
    avg = round(sum(marks) / len(marks), 2)
    status = "Pass" if avg >= 60 else "Fail"
    results.append((name, avg, status))

# Print table
print("\nName            | Average | Status")
print("-----------------------------------")

for r in results:
    print(f"{r[0]:15} | {r[1]:7} | {r[2]}")

# Statistics
pass_count = sum(1 for r in results if r[2] == "Pass")
fail_count = sum(1 for r in results if r[2] == "Fail")

topper = max(results, key=lambda x: x[1])
class_avg = round(sum(r[1] for r in results) / len(results), 2)

print("\nPass:", pass_count)
print("Fail:", fail_count)
print("Topper:", topper[0], topper[1])
print("Class Average:", class_avg)
```
